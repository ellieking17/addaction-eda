```python
import pandas as pd
import glob
import os

dftarget = pd.read_csv ('C:/Users/Robin.Heath/Documents/Machine Learning/Helper Files\DecodingLists.csv')#read in target field list csv
path = 'C:/Users/Robin.Heath/Documents/Machine Learning/Output'#set path for csv extracts to be picked up from
output = 'C:/Users/Robin.Heath/Documents/Machine Learning/OutputDecoded/'#set path for output files to be saved
decode_path = 'C:/Users/Robin.Heath/Documents/Machine Learning/ReferenceLists'#set path for reference list extracts to be picked up from

allfiles = glob.glob(path + '/*.csv')#define complete path/type for file loop
path, dirs, files = next(os.walk(path))#define folder to count files in
file_count = len(files)#capture file count

def right(s, amount):#right function
    return s[-amount:]

bln_open = False#set boolean value
xcount = 0#set counter value

for file_ in allfiles:#loop through target folder
    
    filename = right(file_,len(file_)-len(path)-1)#create file name removing path prefix
    xcount = xcount+1#increment file loop counter
    
    for index, row in dftarget.iterrows():#loop through dftarget fields csv rows
        
        targetfile = row['File']#capture file name
        targetfield = row['Field']#capture field name
        decodefile = row['DecodingList']#capture decoding file name 
            
        if targetfile == filename:#check if target file matches with current file

            if bln_open == False:#check if boolean var is false (matching file not opened yet)
                
                df = pd.read_csv(file_,low_memory=False,encoding = "ISO-8859-1")#read in current file to dataframe
                bln_open = True#change boolean var to true (no need to open file again)
            
            for column in df:#loop through current file columns
                
                sourcecol = column#capture column head value
                
                if targetfield == sourcecol:#check if dftarget target field matches with df column head

                    df_decoder = pd.read_csv(decode_path + decodefile,delimiter ='|',encoding = "ISO-8859-1")#read decode file into dataframe
                    
                    x = df_decoder['Code'].count()#capture the number of rows within the target column

                    z = targetfield#pass column head

                    for x in range (x):#loop through populated rows in target file, target column
                        
                        c_val = df_decoder['Code'].loc[x]#capture target column row value
                        t_val = df_decoder['Text'].loc[x]#capture target column row value

                        df[z].replace(to_replace=c_val,value=t_val,inplace=True)        
                            
        else:#if targetfile does not match with current file
            
            if bln_open == True:#check if boolean is true
                df.to_csv(output + filename,index=False)#write the dataframe to csv
                
                bln_open = False#change boolean var to false 
                
    if xcount == file_count:#check if on last file in loop
        
        if bln_open == True:#check if boolean is true
            
            df.to_csv(output + filename,index=False)#write the dataframe to csv
        
                    
                    
```
